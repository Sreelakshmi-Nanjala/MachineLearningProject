## Selecting the features for model building

```python
x_train_df=x_train[['Total_Trans_Ct','Total_Revolving_Bal',
'Total_Relationship_Count','Total_Ct_Chng_Q4_Q1','Months_Inactive_12_mon']]
x_test_df=x_test[['Total_Trans_Ct','Total_Revolving_Bal','Total_Relationship_Count',
                  'Total_Ct_Chng_Q4_Q1','Months_Inactive_12_mon']]
```

**Checking the normal distribution for the numerical features and transform them**

<div><b>Below method would check numerical features are noamally distributed or not, based on that we are going to apply scalar to the features</b></div>

```python
def test_normality(df):
    normality_results = {}
    
    for column_name in df.columns:
        reject_h0 = False

        data = df[column_name]
        nordistest = anderson(data)

        for i in range(len(nordistest.critical_values)):
            sl, cv = nordistest.significance_level[i], nordistest.critical_values[i]
            if nordistest.statistic >= cv:
                reject_h0 = True
                break

        if reject_h0:
            conclusion = f"Data is not normally distributed (reject H0)"
        else:
            conclusion = f"Data is normally distributed (fail to reject H0)"
        
        normality_results[column_name] = conclusion

    return normality_results
```

```python
normality_results = test_normality(x_train_df)
for column_name, result in normality_results.items():
    print(f"Normality: '{column_name}': {result}")
```

</div><b> Applying the Standard scalar to the features before sending data to the model</b></div>

```python
scaler = MinMaxScaler()
x_max_train_scaled = scaler.fit_transform(x_train_df)
x_max_test_scaled = scaler.transform(x_test_df)
```

# Hyperparameter Tuning for the models

#### Random Forest tuning

```python
rfparameters= {
    'n_estimators':[200,400],
    'max_features':['sqrt','log2'],
    'max_depth':[10,20,30,40],
    'criterion':['gini','entropy']}
rfclf= RandomForestClassifier()
rand_rfclf = RandomizedSearchCV(rfclf, rfparameters, cv = 5, verbose=True, n_jobs=2)
rand_rfclf.fit(x_max_train_scaled,y_train)
print(rand_rfclf.best_params_)
```

#### XGboost Tuning

```python
xgbparams = {
    "learning_rate": [0.01, 0.1, 0.2],
    "max_depth": [3, 5, 7, 9],
    "n_estimators": [50, 100, 150, 200],
    "min_child_weight": [1, 3, 5],
    "subsample": [0.8, 0.9, 1.0],
    "colsample_bytree": [0.8, 0.9, 1.0],
    "gamma": [0, 1, 5]
}

xgbclf= XGBClassifier()
xgb_clf = RandomizedSearchCV(xgbclf, xgbparams, cv = 5, verbose=True, n_jobs=2)
xgb_clf.fit(x_max_train_scaled,y_train)
print(xgb_clf.best_params_)
```

#### Adaboost Tuning

```python
adaparams= {
            'n_estimators': [10, 50, 100, 200,300],
            'learning_rate': [0.01, 0.1, 0.5, 1],
        }

adaclf=AdaBoostClassifier()
ada_boostclf = RandomizedSearchCV(adaclf, adaparams, cv = 5, verbose=True, n_jobs=-1)
ada_boostclf.fit(x_max_train_scaled,y_train)
print(ada_boostclf.best_params_)
```
